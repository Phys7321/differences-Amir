```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0.1,1.9,0.01)
```

```python
print (x)
```

```python
y= lambda x: (np.sin(1/(x*(2-x))))**2
```

```python
plt.plot (x,y(x),label='y')
plt.plot (x,np.gradient(y(x),1),label='dy/dx')
plt.plot (x,np.gradient(y(x),2),label='d2y/dx2')
```
