```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import array
x1=np.arange(-2,-0.01,0.01)
x2=np.arange(0.01,1.99,0.01)
x3=np.arange(2.01,4,0.01)
x=np.append(x1,x2)
x=np.append(x,x3)
```

```python
N = len(x)
f = lambda x: np.sin(1/(x*(2-x)))**2
vectorf = np.vectorize(f)
```

```python
plt.figure()
plt.plot(x,vectorf(x))
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
a=-2
b=4
def centraldiff(f,a,b,N):
    g = []
    h = (b-a)/N
    for k in range(0,N):
        slop = (f(a+(k+1/2)*h)-f(a+(k-1/2)*h))/h
        g.append(slop)
    return g
```

```python
cdy = []

cdy = array(centraldiff(f,a,b,N))
```

```python
plt.figure()
plt.plot(x,cdy)
plt.xlabel('x')
plt.ylabel('df/dx')
```

```python
a=-1
b=3

def centraldiff_2ndDer (f,a,b,N):
    g=[]
    h = (b-a)/N
    
    for k in range(0,N):
        slop = (f(a+(k+1)*h)+f(a+(k-1)*h)-2*f(a+k*h))/(h**2)
        g.append(slop)
    return g
```

```python
cd2y = []

cd2y = array(centraldiff_2ndDer(f,a,b,N))
```

```python
plt.figure()
plt.plot(x,cd2y)
plt.xlim(-1,3)
plt.xlabel('x')
plt.ylabel('d2f/dx2')
```
